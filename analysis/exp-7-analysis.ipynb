```python
import pandas as pd

def clean_data(df):
    # Drop columns: 'data', 'epochs' and 12 other columns
    df = df.drop(columns=['horizon','experiment_name', 'tokenization', 'epochs', 'batch_size', 'learning_rate', 'test_split', 'val_split', 'vocab_size', 'patience', 'results_dir', 'shuffle', 'disable_train_shuffle', 'early_stopping', 'remove_system_tables'])
    # Sort by column: 'actual_test_accuracy_mean' (descending)
    #df = df.sort_values(['horizon'], ascending=[True])
    return df

df = pd.read_csv('../results/exp-5-large/results_clean.csv')
df_clean = clean_data(df.copy())

df_longer = pd.read_csv('../results/exp-7/results_clean.csv')
df_clean_longer = clean_data(df_longer.copy())
```

```python

short = df_clean[df_clean['add_start_end_tokens'] & df_clean['add_row_id'] & df_clean['add_label_tokens'] & (df_clean['train_data_percent_used'] == 1.0)]
long = df_clean_longer[df_clean_longer['add_start_end_tokens'] & df_clean_longer['add_row_id'] & df_clean_longer['add_label_tokens'] & (df_clean_longer['train_data_percent_used'] == 1.0)]

concat = pd.concat([short, long])
concat
```
