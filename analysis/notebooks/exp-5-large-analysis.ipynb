```python
import pandas as pd

def clean_data(df):
    # Drop columns: 'data', 'epochs' and 12 other columns
    df = df.drop(columns=['data', 'epochs', 'batch_size', 'learning_rate', 'seq_length', 'test_split', 'val_split', 'vocab_size', 'patience', 'results_dir', 'shuffle', 'disable_train_shuffle', 'early_stopping', 'remove_system_tables'])
    # Sort by column: 'actual_test_accuracy_mean' (descending)
    df = df.sort_values(['actual_test_accuracy_mean'], ascending=[False])
    return df

df = pd.read_csv('../results/exp-5-large/results_clean.csv')

df_clean = clean_data(df.copy())
```

```python
# for each train_data_percent_used, get the row with the highest actual_test_accuracy_mean
df_best_per_percent = df_clean.loc[df_clean.groupby('train_data_percent_used')['actual_test_accuracy_mean'].idxmax()]
df_best_per_percent.sort_values('train_data_percent_used', ascending=[False], inplace=True)
df_best_per_percent
```
